```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizations ko notebook me hi inline show karne ke liye
%matplotlib inline
print("Libraries successfully imported!")
```

```python
df = pd.read_excel('C:/Users/abcd/Desktop/python project/ecommerce_sales_data.xlsx', sheet_name='Sales Data')
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df["OrderDate"].dtype
```

```python
total_sales = df['Amount'].sum()
total_profit = df['Profit'].sum()
avg_quantity = df['Quantity'].mean()
print("===== BUSINESS OVERVIEW =====")
print(f"Total Revenue Generated: ₹{total_sales:,}")
print(f"Total Profit Earned: ₹{total_profit:,}")
print(f"Average Items Per Order: {avg_quantity:.1f}\n")
```

```python
# 2. Category-wise Performance (Group By)
print("===== CATEGORY PERFORMANCE =====")
category_perf = df.groupby('Category').agg({'Amount': 'sum', 'Profit': 'sum', 'Quantity': 'sum'}).reset_index()
print(category_perf)
```

```python
# 3. State-wise Top Revenue (Ranking)
print("\n===== TOP REVENUE STATES =====")
state_perf = df.groupby('State')['Amount'].sum().sort_values(ascending=False).reset_index()
print(state_perf)
```

```python
# Plot 1: State-wise Revenue Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(x='Amount', y='State', data = state_perf, palette='Blues_r')
plt.title('State-wise Revenue Breakdown')
plt.xlabel('Total Sales (₹)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Category-wise Amount vs Profit
plt.figure(figsize=(8, 4))
sns.barplot(x='Category', y='Amount', data=category_perf, color='skyblue', label='Total Sales')
sns.barplot(x='Category', y='Profit', data=category_perf, color='navy', label='Total Profit')
plt.title('Sales vs Profit by Product Category')
plt.ylabel('Value (₹)')
plt.legend()
plt.tight_layout()
plt.show()
```
